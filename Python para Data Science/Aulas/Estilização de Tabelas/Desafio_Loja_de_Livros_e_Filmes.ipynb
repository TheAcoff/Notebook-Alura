Foi proposto o desafio para trabalhar com os dados para fazer tabelas que podem ser adequadas para diversas análises para essa loja.

```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/DV-estilizacao-tabelas/main/Atividade%20de%20desafio/loja_livro_filmes.csv'
dados = pd.read_csv(url)
dados
```

## Desafio 1
Temos que analisar quais países mais solicitaram produtos por pedidos. Com isso, podemos fazer um estudo quanto a distribuição e logística dos produtos.

```python
paises_produtos = dados.groupby(['pais'])['unidades'].sum().nlargest(10).copy()
paises_produtos = paises_produtos.reset_index()
paises_produtos.columns = ['País','Unidades']
paises_produtos
```

```python
paises_produtos['Rank'] = paises_produtos.index+1
paises_produtos.set_index('Rank', inplace = True)
paises_produtos
```

```python
s_pais = paises_produtos.style
s_pais = s_pais.format({'Unidades': '{} produtos'})
s_pais
```

## Desafio 2

Construir uma visualização que permita identificar quais categorias de produto geraram mais e menos custos para a loja.

```python
custos = dados.groupby(['categoria'])['custo_produto'].sum().copy()
custos = custos.reset_index()
custos.columns = ['Categoria','Custo']
custos.set_index('Categoria', inplace = True)
custos
```

Transformando e formatando o style para apresentar o maior e o menor valor de forma destacada

```python
s_custos = custos.style
s_custos.format('R$ {:,.2f}')\
        .highlight_max(color = 'red')\
        .highlight_min(color = 'green')
s_custos
```

Formatando o cabeçalho

```python
cabecalho = {
    'selector':'th', #Selecionar qual parte da tabela vai ser estilizada (th -> header)
    'props':'font-weight: bold; font-family: Arial; text-align: center; text-transform: capitalize;' #Propriedades (fonte, alinhamento, etc.)}
}
s_custos.set_table_styles([cabecalho], overwrite = False)
```

## Desafio 3
Informar a quantidade e distribuição de pedidos por tipo de desconto.

```python
descontos = pd.DataFrame(dados['tipo_desconto'].value_counts())
descontos.columns = ['N° Pedidos']
descontos.index.name = 'Tipo desconto'

descontos
```

Criando uma visualização de porcentagem

```python
import numpy as np

porcentagem = descontos['N° Pedidos'].to_numpy()
porcentagem = 100*porcentagem/porcentagem.sum()
porcentagem
```

```python
descontos['Porcentagens'] = porcentagem
descontos
```

```python
s_descontos = descontos.style

cabecalho = {
    'selector': 'th',
    'props': 'font-weight: bold; font-family: Arial; text-align: right; background-color: white; color: black;'
}
celulas = {
    'selector': 'td',
    'props': 'background-color: white; color: black;'
}

s_descontos.set_table_styles([cabecalho,celulas], overwrite=False)
s_descontos
```

```python
s_descontos.format({'Porcentagens': '{:.2f} %'})\
           .bar(subset='Porcentagens', vmin = 0, vmax = 100.0, color = '#093364',
                height=50,width=60)

s_descontos
```
