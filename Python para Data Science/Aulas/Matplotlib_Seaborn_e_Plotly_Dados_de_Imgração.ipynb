# Matplotlib

```python
import pandas as pd
```

```python
uri = 'https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/main/Dados/imigrantes_canada.csv'
```

```python
dados = pd.read_csv(uri)
dados.head()
```

> Descrever a tabela


```python
dados.info()
```

Mudamos o índice

```python
dados.set_index('País', inplace = True)
```

Criando uma variável para armazenar todo o intervalo de tempo

```python
anos = list(map(str, range(1980,2014)))
```

## Dados do Brasil
Vamos separar as informações com os dados do Brasil para visualizar.

```python
brasil = dados.loc['Brasil',anos]
brasil_dict = {'ano':brasil.index.tolist(), 'imigrantes':brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil
```

Criando um gráfico com o matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (8,4))
plt.title('Número de Imigrantes por Ano no Brasil')
plt.plot(dados_brasil['ano'],dados_brasil['imigrantes'])
plt.xticks(list(map(str,range(1980,2011,5))))
plt.xlabel('Anos')
plt.ylabel('Imigrantes')
plt.show()
```

---
## Comparando a Tendência de Imigração do Brasil e Argentina
> Desafio proposto para analisar e comparar os dados

Começamos cirando um dataframe que une as informações do Brasil e da Argentina

```python
brasil_argentina = dados.loc[['Brasil','Argentina'],anos].T
brasil_argentina
```

```python
plt.figure(figsize = (8,5))
plt.title('Imigração Brasil e Argentina para o Canadá')
plt.plot(anos,brasil_argentina['Brasil'])
plt.plot(anos,brasil_argentina['Argentina'])
plt.xticks(list(map(str,range(1980,2011,5))))
plt.xlabel('Anos')
plt.ylabel('Imigrantes')
plt.legend(['Brasil','Argentina'])
plt.show()
```

Notamos que no início, as tendências de migração estavam bem próximas. Porém, houve uma separação depois de 2000, com um aumento da Argentina. Logo depois, houve uma queda, seguida do aumento do número de imigrantes do Brasil.

## Maiores países da América do Sul
> Conteúdo da aula

```python
#Definindo a figura e seus eixos
fig, axs = plt.subplots(2,2, figsize = (10,6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)
fig.suptitle('Imigração dos 4 maiores países da América do Sul para o Canadá',fontsize = 18)

#Especificando os gráficos
axs[0,0].plot(dados.loc['Brasil',anos],color='green')
axs[0,0].set_title('Brasil')

axs[0,1].plot(dados.loc['Colômbia',anos],color='orange')
axs[0,1].set_title('Colômbia')

axs[1,0].plot(dados.loc['Argentina',anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(dados.loc['Peru',anos], color = 'red')
axs[1,1].set_title('Peru')

#Automatizando os rótulos
for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5)) #Pular os anos de 5 em  5
  ax.set_xlabel('Ano')
  ax.set_ylabel('Imigrantes')

#Normalizando o eixo y com o número de imigrantes
ymin = 0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymin,ymax)


plt.show()
```

## América do Sul
> Conteúdo da aula

Iremos plotar a quantidade total de imigrações para cada países com um gráfico de barras

```python
america_sul = dados.query('Região == "América do Sul"')

america_sul_sorted = america_sul.sort_values('Total',ascending = True)
america_sul_sorted
```

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown',
         'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize = (12,6))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color = cores)
ax.set_title('Imigração Total da América do Sul para o Canadá \n 1980 a 2013',
             loc = 'left', fontsize = 18)
ax.set_xlabel('Número de Imigrantes', fontsize = 14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize = 12)
ax.yaxis.set_tick_params(labelsize = 12)

plt.show()
```

Dados específicos para o Brasil

```python
cores = []

for pais in america_sul_sorted.index:
  if pais == 'Brasil':
    cores.append('green')
  else:
    cores.append('silver')

fig, ax = plt.subplots(figsize = (12,6))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color = cores)
ax.set_title('Imigração Total do Brasil comparado com América do Sul \n para o Canadá de 1980 a 2013',
             loc = 'left', fontsize = 18)
ax.set_xlabel('Número de Imigrantes', fontsize = 14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize = 12)
ax.yaxis.set_tick_params(labelsize = 12)

#Criando anotações
for i, v in enumerate(america_sul_sorted['Total']):
  ax.text(v, i, str(v), color = 'black', fontsize =10, ha = 'left', va = 'center')

plt.show()
```

# Seaborn

```python
import seaborn as sns
```

```python
sns.set_theme()
```

```python
top_10 = dados.sort_values('Total', ascending=False).head(10)
top_10
```

Criando um gráfico de barras

```python
ax = sns.barplot(data = top_10, x = 'Total', y = top_10.index, orient = 'h')

ax.set(title='Países com maior imigração para o Canadá \n 1980 a 2013',
       xlabel = 'Número de Imigrantes',
       ylabel = '')

plt.show()
```

Misturando com o Matplotlib e definindo paletas

```python
def gerar_gráfico(paleta):
  fig, ax = plt.subplots(figsize = (8,4))
  ax = sns.barplot(data = top_10,
                  x = 'Total', y = top_10.index,
                  orient = 'h', palette=paleta)

  ax.set_title('Países com maior imigração para o Canadá \n 1980 a 2013',
              loc = 'left',
              fontsize = 16)
  ax.set_xlabel('Número de imigrantes', fontsize = 14)
  ax.set_ylabel('')

  plt.show()

gerar_gráfico('tab10')
```

## Explorando outros temas

```python
sns.set_theme(style='dark')
gerar_gráfico('tab10')
```

```python
sns.set_theme(style='whitegrid')
gerar_gráfico('tab10')
```

# Plotly

```python
import plotly.express as px
```

Faremos os gráficos com os dados do Brasil

```python
dados_brasil.head()
```

```python
fig = px.line(dados_brasil, x='ano', y='imigrantes', title='Imigrantes do Brasil para o Canadá')
fig.update_traces(line_color='green', line_width=4)
fig.update_layout(width=1000, height = 500,
                  xaxis={'tickangle':-45},
                  font_family='Arial',
                  font_size=14,
                  font_color='grey',
                  title_font_color='black',
                  title_font_size=22,
                  xaxis_title='Ano',
                  yaxis_title='Número de imigrantes')
fig.show()
```

Multi dados interativo, os 7 maiores números de imigração dos páises da América do Sul

```python
top_7 = america_sul_sorted.tail(7)
top_7 = top_7.drop(['Continente','Região','Total'], axis=1).T

fig = px.line(top_7,
              title = 'Imigração dos Países da América do Sul')
fig.update_layout(
    xaxis = {'tickangle':-45},
    xaxis_title = 'Anos',
    yaxis_title = 'Número de Imigrantes',
)
fig.show()
```

```python
fig.write_html('imigracao.html')
```

# Animando o Gráfico

```python
import plotly.graph_objs as go

dados_brasil['ano'] = dados_brasil['ano'].astype(int)

# Criando uma figura
fig = go.Figure()

# Adicionando a linha do gráfico e definindo a espessura da linha
fig.add_trace(
    go.Scatter(x=[dados_brasil['ano'].iloc[0]], y=[dados_brasil['imigrantes'].iloc[0]], mode='lines', name='Imigrantes', line=dict(width=4))
)

# Definindo as configurações de layout
fig.update_layout(
    title=dict(
        text='<b>Imigração do Brasil para o Canadá no período de 1980 a 2013</b>',
        x=0.12,
        xanchor='left',
        font=dict(size=20)
    ),
    xaxis=dict(range=[1980, 2013], autorange=False, title='<b>Ano</b>'),
    yaxis=dict(range=[0, 3000], autorange=False, title='<b>Número de imigrantes</b>'),
    updatemenus=[dict(
        type='buttons',
        showactive=False,
        buttons=[dict(
            label='Play',
            method='animate',
            args=[None, {'frame': {'duration': 100, 'redraw': True}, 'fromcurrent': True}]
        )]
    )],
    width=1000,
    height=500
)

# Definir as configurações de animação
frames = [go.Frame(data=[go.Scatter(x=dados_brasil['ano'].iloc[:i+1], y=dados_brasil['imigrantes'].iloc[:i+1])]) for i in range(len(dados_brasil))]
fig.frames = frames

# Mostrando a figura
fig.show()
```
