O exercício propõe a criação de um banco de dados de uma instituição financeira usando a base: do [arquivo CSV](https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv).

```python
#Importando as bibliotecas
import pandas as pd
import sqlalchemy
#Atualizando a versão do sqlalchemy para compatbilidade
!pip install --upgrade 'sqlalchemy<2.0'
```

```python
from sqlalchemy import create_engine, MetaData, Table, inspect
```

Importando os dados para um DataFrame e fazendo as primeiras análises.

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv'
df = pd.read_csv(url)
df.head()
```

```python
df.shape #Formato do DataFrame
```

```python
df.info() #Informações das variáveis
```

Verificando se há a existência de valores não numériocs

```python
df.isnull().sum()
```

Criando o banco de dados

```python
engine = create_engine('sqlite:///:memory:')
```

Importando o DataFrame para o Banco de Dados

```python
df.to_sql('clientes', engine, index=False)
```

Verificando na base de dados se há o df

```python
inspector = inspect(engine)
inspector.get_table_names()
```

```python
querry = 'SELECT * FROM clientes'
pd.read_sql(querry,engine)
```

---
Com a base de dados criada, é pedida algumas atualizações:
1. Atualizar o registro do cliente de ID 6840104 que teve o rendimento anual alterado para 300000.

```python
querry = 'UPDATE clientes SET Rendimento_Anual = 300000 WHERE ID_Cliente = 6840104'
with engine.connect() as conn:
	conn.execute(querry)
```

Conferimos o registro do cliente 6840104

```python
querry = 'SELECT * FROM clientes WHERE ID_Cliente = 6840104'
pd.read_sql(querry,engine)
```

2. Excluir o registro do cliente de ID 5008809, pois essa pessoa não possui mais conta na instituição financeira.

```python
querry = 'DELETE FROM clientes WHERE ID_Cliente = 5008809'
with engine.connect() as conn:
	conn.execute(querry)
```

```python
querry = 'SELECT * FROM clientes WHERE ID_Cliente = 5008809'
pd.read_sql(querry,engine)
```
