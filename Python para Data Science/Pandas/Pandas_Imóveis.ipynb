Filtragem de Dados: tirando dados comerciais

```python
dados.Tipo.unique()
```

```python
imoveis_comerciais = ['Conjunto Comercial/Sala',
                      'Prédio Inteiro', 'Loja/Salão',
                      'Galpão/Depósito/Armazém',
                      'Casa Comercial', 'Terreno Padrão',
                      'Loja Shopping/ Ct Comercial',
                      'Box/Garagem', 'Chácara',
                      'Loteamento/Condomínio', 'Sítio',
                      'Pousada/Chalé', 'Hotel', 'Indústria']
```

```python
df_residencial = dados.query('@imoveis_comerciais not in Tipo')
df_residencial.head()
```

```python
df_residencial.Tipo.unique()
```

```python
df_preco_tipo = df_residencial.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_preco_tipo.plot(kind='barh',figsize=(14,10),color='blue')
```

---
Porcentagem

```python
df_percentual = df_residencial.Tipo.value_counts(normalize=True).to_frame().sort_values('proportion')
df_percentual
```

```python
df_percentual.plot(kind='bar',figsize=(14,10),color='blue',xlabel = 'Tipos',ylabel = 'Porcentagem')
```

# Tratando Dados Nulos

```python
df.isnull().sum()
```

```python
df = df.fillna(0)
df.isnull().sum()
```

Remover apartamentos com Valor = 0 e também com condomínio = 0

```python
indices_remover = df.query('Valor == 0 | Condominio == 0').index
df.drop(indices_remover, axis=0, inplace=True)
df.head()
```

Remover coluna 'Tipo' com Apartamentos

```python
df.drop('Tipo',axis=1,inplace=True)
df.head()
```

---
# Aplicar Filtros



## 1. Apartamentos que possuem 1 quarto e aluguel menor que R$ 1200;


```python
df['Quartos'] == 1
```

```python
filtro1 = df['Quartos'] == 1
df[filtro1]
```

```python
filtro2 = df['Valor'] < 1200
df_1 = df[filtro1 & filtro2]
```

## 2. Apartamentos que possuem pelo menos 2 quartos, aluguel menor que R$ 3000 e área maior que 70 m².

```python
filtro = (df['Quartos'] >= 2) & (df['Valor'] < 3000) & (df['Area'] > 70)
df_2 = df[filtro]
```

Salvando os arquivos

```python
df_1.to_csv('df_1.csv', sep=';', index=False)
df_2.to_csv('df_2.csv', sep=';', index=False)
```
